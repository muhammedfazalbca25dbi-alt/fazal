Task:- 3 Check and Remove Duplicate

```python
print("Duplicate Rows:",df.duplicated().sum())
dupes = df[df.duplicated(subset=["Order ID"], keep=False)]
print(dupes.shape)
dupes.head()
```

Task:- 4 Convert Date Coloumns

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
(df.dtypes)
```

Task:- 5 Engineer Shipping Days

```python
df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days
(df.groupby("Region")["Shipping Days"].mean().round(1))
```

Task:- 7 Engineer Profit Margin

```python
df["Profit Margin"] = (df["Profit"] / df["Sales"] * 100).round(2)
(df.groupby("Sub-Category")["Profit Margin"].mean()
    .sort_values(ascending=True).round(1))
```

Task:- 6 Engineer Order Year

```python
df["Order Year"] = df["Order Date"].dt.year
```
